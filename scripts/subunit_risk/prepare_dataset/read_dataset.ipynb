输出查看对话格式的数据集

```python
from datasets import load_dataset
from PIL import Image
import matplotlib.pyplot as plt

# 加载数据集
dataset = load_dataset("/root/group-shared/yaww-ai-lesson/xwz/code/RisKVA/datasets/RisKVA/Subunit-Risk_QA_v1")["train"]

print("dataset：", dataset)
```

```python
for i in range(20):
    print("=" * 30)

    print("⭐️ 显示 dataset：", dataset[i*6])

    print("⭐️ 显示文本长度：", len(dataset[i*6]["messages"]))

    msg = dataset[i*6]["messages"][0]
    print("⭐️ 显示 user 文本：", msg)
    msg = dataset[i*6]["messages"][1]
    print("⭐️ 显示 assistant 文本：", msg)

    print("⭐️ 显示图片数量：", len(dataset[i*6]["images"]))

    img = dataset[i*6]["images"][0]
    print("⭐️ 显示图片：", img)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
